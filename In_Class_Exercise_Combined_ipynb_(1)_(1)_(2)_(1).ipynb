### Q2. Apply VADER to compute sentiment scores (3 points)
- Import VADER
- Download the lexicon if needed
- Create a new column called `compound`
- Show the first 5 rows of `Text` and `compound`

```python
!pip -q install nltk
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df['compound'] = df['Text'].apply(lambda text: sia.polarity_scores(text)['compound'])

print(df[['Text', 'compound']].head())
```

### Q3. Create a sentiment label from the compound score (3 points)
Create a new column called `PredictedSentiment` using this rule:
- `compound >= 0.05` → `positive`
- `compound <= -0.05` → `negative`
- otherwise → `neutral`

Then show the first 10 rows of `Text`, `compound`, and `PredictedSentiment`.

```python
# Write your answer here
def label_sentiment(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['PredictedSentiment'] = df['compound'].apply(label_sentiment)

print(df[['Text', 'compound', 'PredictedSentiment']].head(10))
```

### Q4. Compare review score and sentiment (2 points)
Group by `Score` and compute the **average compound score** for each score value.

```python
# Write your answer here
avg_compound = df.groupby('Score')['compound'].mean().reset_index()
avg_compound.columns = ['Score', 'AvgCompound']
print(avg_compound)
```

### Q1. Zero-shot classification prompt (3 points)
Write a **zero-shot prompt** that asks an AI model to classify each review as **positive, negative, or neutral**.

Your prompt should include:
- a clear instruction
- the three labels
- a request for a clean output format

Store your prompt in a Python variable called `zero_shot_prompt`, then print it.

```python
# Write your answer here
zero_shot_prompt = """You are a sentiment classifier. Classify each review below as exactly one of: positive, negative, or neutral.

Return only the label for each review, one per line, in the format:
Review 1: <label>
Review 2: <label>
Review 3: <label>

Reviews:
1. The product arrived late and the quality was poor.
2. Amazing battery life and elegant design.
3. It is fine for daily use, nothing special.
"""

print(zero_shot_prompt)
```

### Q2. Few-shot prompt improvement (3 points)
Now improve the previous prompt by adding **two examples**.

Requirements:
- Store it in a variable called `few_shot_prompt`
- Include 2 input-output examples
- Ask the model to classify the 3 sample reviews after the examples

```python
# Write your answer here
few_shot_prompt = """You are a sentiment classifier. Classify each review as exactly one of: positive, negative, or neutral.

Here are two examples:

Example 1:
Review: "The delivery was fast and the product exceeded my expectations."
Label: positive

Example 2:
Review: "The item broke after one use. Very disappointed."
Label: negative

Now classify the following reviews:
Review 1: "The product arrived late and the quality was poor."
Review 2: "Amazing battery life and elegant design."
Review 3: "It is fine for daily use, nothing special."

Return your answer in this format:
Review 1: <label>
Review 2: <label>
Review 3: <label>
"""

print(few_shot_prompt)
```

### Q3. Prompt comparison and refinement (4 points)
Create a short table or DataFrame with **three columns**:
- `PromptType`
- `MainFeature`
- `WhyUseful`

Include these three prompt types:
1. Zero-shot
2. Few-shot
3. Chain-of-thought

Then, in **2–3 sentences**, explain which one you would choose for a more difficult text analysis task and why.

```python
# Write your answer here
import pandas as pd

comparison_data = {
    'PromptType': ['Zero-shot', 'Few-shot', 'Chain-of-thought'],
    'MainFeature': [
        'Instruction only, no examples',
        'Instruction with 2+ labeled examples',
        'Model reasons step-by-step before answering'
    ],
    'WhyUseful': [
        'Fast to write; works well for simple, clear-cut tasks',
        'Improves accuracy by showing the model the expected format and label style',
        'Best for ambiguous or multi-step tasks where reasoning reduces errors'
    ]
}

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df.to_string(index=False))

print("""
For a more difficult text analysis task — such as detecting sarcasm or
classifying reviews with mixed signals — I would choose chain-of-thought
prompting. It forces the model to reason through the text before committing
to a label, which reduces confident wrong answers on ambiguous inputs. Few-shot
helps with format consistency, but without explicit reasoning steps the model
can still misclassify edge cases.
""")
```
